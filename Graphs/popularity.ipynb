```python
import matplotlib.pyplot as plt
```

```python
athletes = ['Federico Chiesa', 'Matteo Berrettini', 'Gianluigi Donnarumma', 'Lorenzo Insigne',\
    'Ciro Immobile', 'Leonardo Spinazzola', 'Gianluigi Buffon', 'Giorgio Chiellini', 'Jannik Sinner',\
        'Dorothea Wierer', 'Sofia Goggia', 'Alessio Romagnoli', 'Alessandro Florenzi', 'Claudio Marchisio',\
            'Ignazio Moser', 'Lenardo Bonucci', 'Andrea Pirlo', 'Ivan Zaytsev', 'Lorenzo Pellegrini',\
                'Gregorio Paltrinieri']


social_media_value = [100.132, 49.342, 40.507, 25.556, 19.361, 17.124, 16.934, 16.031, 14.377,\
    12.367, 12.364, 11.344, 10.263, 9.931, 7.864, 7.825, 7.833, 6.912, 6.809, 6.711]                 
```

```python
def make_rgb_transparent(rgb, bg_rgb, alpha):
    return [alpha * c1 + (1 - alpha) * c2
            for (c1, c2) in zip(rgb, bg_rgb)]
```

```python
fig, ax = plt.subplots(1, 1, figsize = (15, 9))
ax.barh(athletes[::-1], social_media_value[::-1], alpha = 1, edgecolor = 'black',\
    linewidth = 2, color = 'firebrick', zorder = 2, label= 'Social Media Value (valori medi)')
ax.set_yticks(range(len(athletes)))
ax.set_yticklabels(athletes[::-1])
ax.minorticks_on()
ax.tick_params(axis='both', which='major', labelsize=15, length=15)
ax.tick_params(axis='x', which='minor', labelsize=15, length=8)
ax.tick_params(axis='y', which='minor', length=0)
ax.grid(zorder = 1)
ax.legend(facecolor = 'aliceblue', edgecolor = 'black', shadow = True, fontsize = 20)
for i, txt in enumerate(social_media_value[::-1]):
    ax.text(txt + 1.500, i-0.25, str(txt), fontsize = 15,\
        bbox=dict(facecolor='white', edgecolor= 'white', boxstyle='round, pad=0'))
ax.set_xlim(left = 0, right = 111.132)
ax.set_title('G' + 'LI INFLUENCER SPORTIVI ITALIANI CON PI ALTO VALORE MEDIATICO'.lower(),\
    fontsize = 25)
fig.savefig('../../../../Desktop/popularity.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```
