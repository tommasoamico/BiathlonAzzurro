```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as mtick
```

```python
perc_ind_m = 0.7777778
perc_oth_m = 0.813253
perc_ind_w = 0.7951807
perc_oth_w = 0.8209877
```

```python
width=0.3
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
ax.grid(zorder = 1, alpha = 0.8)
ax.bar(x = [1, 2], height = [perc_ind_m*100, perc_ind_w*100], width=width, label = 'Individuale',\
    color = 'navy', alpha = 0.8, edgecolor = 'black', linewidth = 3, zorder = 2)
ax.bar(x = np.array([1, 2]) + width, height = [perc_oth_m*100, perc_oth_w*100], width=width,\
     label = 'Inseguimento e Mass Start', color = 'firebrick', alpha = 0.8, edgecolor = 'black',\
        linewidth = 3, zorder = 2)
ax.text(x = 1-0.06, y = perc_ind_m*100 + 1.5, s = f'{perc_ind_m*100:.2f}%',\
    fontdict={'fontsize':20}, zorder = 2)
ax.text(x = 2-0.06, y = perc_ind_w*100 + 1.5, s = f'{perc_ind_w*100:.2f}%',\
    fontdict={'fontsize':20}, zorder = 2)
ax.text(x = 1-0.06 + width, y = perc_oth_m*100 + 1.5, s = f'{perc_oth_m*100:.2f}%',\
    fontdict={'fontsize':20}, zorder = 2)
ax.text(x = 2-0.06 + width, y = perc_oth_w*100 + 1.5, s = f'{perc_oth_w*100:.2f}%',\
    fontdict={'fontsize':20}, zorder = 2)
ax.set_ylim(top=100, bottom = 0)
ax.minorticks_on()
ax.tick_params(axis='both', which='major', labelsize=15, length=15)
ax.tick_params(axis='x', which='major', labelsize=15, length=15)
ax.tick_params(axis='y', which='minor', labelsize=15, length=8)
ax.tick_params(axis='x', which='minor', length=0)
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
ax.legend(facecolor = 'aliceblue', edgecolor = 'black', shadow = True, fontsize = 15)
ax.set_ylabel('Percentuale di bersagli colpiti', fontsize = 20)
ax.set_title('Percentuali dell\'ultimo bersaglio dopo un $19/19$', fontsize = 25)
ax.set_xticks([1 + width/2, 2 + width/2])
ax.set_xticklabels(['Uomini', 'Donne'], fontsize = 20)
plt.show()
```
