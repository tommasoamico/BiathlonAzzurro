```python
# Load the image
image = cv2.imread("../../../Desktop/Lisa.png")
reby_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
reby_rgb[reby_rgb == 0] = 255
plt.imshow(reby_rgb)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 8))
edges = cv2.Canny(image, 100, 200)
edges[edges > 190] = 255
edges[150:280, 345:450] = 0
edges[150:200, 320:460] = 0
ax.imshow(edges, cmap = 'viridis')
ax.axis('off')
```
