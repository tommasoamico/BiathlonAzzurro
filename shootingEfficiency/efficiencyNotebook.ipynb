```python
import pandas as pd
import numpy as np
```

```python
rangeTimes = pd.read_csv('./data/rangeTimes.txt', sep='\t')
rangeTimes['fullName'] = rangeTimes['Given Name'] + ' ' + rangeTimes['Family\xa0Name']
rangeTimes.head()
```

```python
shooting = pd.read_csv('./data/shooting.txt', sep='\t')
shooting['fullName'] = shooting['Given Name'] + ' ' + shooting['Family\xa0Name']
shooting['averageErrors'] = (1 - np.array(shooting['Total\nhit rate\n(in %)']/100)) * 5
shooting.head()
```

```python
skiing = pd.read_csv('./data/skiing.txt', sep='\t')
skiing['fullName'] = skiing['Given Name'] + ' ' + skiing['Family\xa0Name']
skiing.head()
```

```python
skiing['back from\naverage\n(in %)'] = skiing['back from\naverage\n(in %)'] - np.min(skiing['back from\naverage\n(in %)'])
skiing.head()
```

```python
jtbPenaltyLoop = 20.8
skiing['averagePenaltyLoop'] = jtbPenaltyLoop + (skiing['back from\naverage\n(in %)']/100) * jtbPenaltyLoop
skiing.head()
```

```python
firstMerge = pd.merge(rangeTimes, shooting, on='fullName')

secondMerge = pd.merge(firstMerge, skiing, on='fullName')
secondMerge = secondMerge.sort_values(by='fullName')
secondMerge = secondMerge.reset_index(drop=True)
secondMerge = secondMerge[['fullName', 'averageErrors', 'averagePenaltyLoop', 'Range\nTime avg\n(in sec)', 'Races', 'Total\nhit rate\n(in %)']]
secondMerge = secondMerge[secondMerge['Races'] > 6]
secondMerge = secondMerge.reset_index(drop=True)
secondMerge.head()
```

```python
secondMerge['timeTotal'] = (secondMerge['averagePenaltyLoop'] * secondMerge['averageErrors']) + secondMerge['Range\nTime avg\n(in sec)']
secondMerge = secondMerge.sort_values(by='timeTotal')
secondMerge = secondMerge.reset_index(drop=True)
secondMerge['Efficiency'] = (100 - np.array(secondMerge['timeTotal'])) / np.max(100 - np.array(secondMerge['timeTotal']))
secondMerge
```

```python
secondMerge[['fullName', 'Efficiency']][:10].to_html()
```
